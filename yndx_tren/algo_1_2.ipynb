## B. Определить вид последовательности

```python
def determine_sequence_type():
    # Инициализация флагов
    ascending = True
    weakly_ascending = True
    descending = True
    weakly_descending = True
    constant = True

    # Чтение первого числа
    prev = int(input())
    if prev == -2 * 10**9:  # Если первое число уже является конечным символом
        print("CONSTANT")
        return

    # Обработка остальных чисел
    while True:
        current = int(input())
        if current == -2 * 10**9:
            break

        # Проверка условий для каждого типа последовательности
        if current > prev:
            descending = False
            weakly_descending = False
            constant = False
        elif current < prev:
            ascending = False
            weakly_ascending = False
            constant = False
        else:  # current == prev
            ascending = False
            descending = False

        prev = current  # Обновление предыдущего значения

    # Определение типа последовательности
    if constant:
        print("CONSTANT")
    elif ascending:
        print("ASCENDING")
    elif weakly_ascending:
        print("WEAKLY ASCENDING")
    elif descending:
        print("DESCENDING")
    elif weakly_descending:
        print("WEAKLY DESCENDING")
    else:
        print("RANDOM")

# Вызов функции
determine_sequence_type()
```

## F. Симметричная последовательность

```python
def is_palindrome(lst):
    return lst == lst[::-1]


def make_palindrome(lst):

    if is_palindrome(lst):
        return 0, []
    
    for i in range(len(lst)):
        if is_palindrome(lst[i:]):  # Находим самый длинный палиндромичный суффикс
            return i, lst[:i][::-1]  # Остаток дописываем в перевернутом порядке
    
    return len(lst) - 1, lst[:-1][::-1]  # В крайнем случае дописываем всю строку наоборот

n = int(input())
lst = list(map(int, input().split()))

m, add_numbers = make_palindrome(lst)

print(m)
if m > 0:
    print(*add_numbers)
```

## G. Наибольшее произведение двух чисел

```python
def max_product_pair(lst):
    min1 = min2 = float('inf')   # Два наименьших числа
    max1 = max2 = float('-inf')  # Два наибольших числа

    for num in lst:
        if num > max1:
            max2, max1 = max1, num
        elif num > max2:
            max2 = num

        if num < min1:
            min2, min1 = min1, num
        elif num < min2:
            min2 = num

    if max1 * max2 > min1 * min2:
        return sorted([max1, max2])
    else:
        return sorted([min1, min2])


lst = list(map(int, input().split()))

print(*max_product_pair(lst))
```

## H. Наибольшее произведение трех чисел

```python
def max_product_triplet(lst):
    min1 = min2 = float('inf')   # Два наименьших числа
    max1 = max2 = max3 = float('-inf')  # Три наибольших числа

    for num in lst:
        # Обновляем три наибольших
        if num > max1:
            max3, max2, max1 = max2, max1, num
        elif num > max2:
            max3, max2 = max2, num
        elif num > max3:
            max3 = num

        # Обновляем два наименьших
        if num < min1:
            min2, min1 = min1, num
        elif num < min2:
            min2 = num

    # Выбираем лучший вариант
    if max1 * max2 * max3 > min1 * min2 * max1:
        return max1, max2, max3
    else:
        return min1, min2, max1


# Ввод данных
lst = list(map(int, input().split()))

# Вывод результата
print(*max_product_triplet(lst))
```

## J. Треугольник Максима

```python
k = int(input())
left = 30.0
right = 4000.0


prev = float(input())
for i in range(k-1):
    now, definer = input().split(' ')
    now = float(now)
  
    if abs(now - prev)<1e-6:
        continue
    if definer == 'closer':
        if now > prev:
            left = max(left, (now + prev)/2)
        else:
            right = min(right, (now + prev)/2)
    else:
        if now < prev:
            left = max(left, (now + prev)/2)
        else:
            right = min(right, (now + prev)/2)
    prev = now

print(left,right)
```
