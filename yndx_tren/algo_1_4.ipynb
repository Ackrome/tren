```python
import re
from collections import defaultdict

# Чтение входных данных
try:
    with open('input.txt', 'r') as f:
        text = f.read()
except FileNotFoundError:
    text = input()

# Разбиение текста на слова с использованием регулярных выражений
words = re.findall(r'\S+', text)

# Словарь для подсчета количества появлений слов
word_count = defaultdict(int)


# Обработка каждого слова
for word in words:
    # Увеличиваем счетчик для этого слова
    word_count[word] += 1

mx = max(word_count.values())
result = []
for k,v in word_count.items():
    if word_count[k] == mx:
        result.append(k)


# Вывод результата
output = sorted(result)[0]
try:
    with open('output.txt', 'w') as f:
        f.write(output)
except:
    print(output)
```

```python
# Чтение входных данных
try:
    with open('input.txt', 'r') as f:
        text = f.read()
except FileNotFoundError:
    text = input()

result = []
text = text.split('\n')
n = int(text[0])
c = list(map(int,text[1].split()))
p = list(map(int,text[3].split()))

# Подсчет количества нажатий для каждой клавиши
press_count = [0] * n  # Создаем список для подсчета нажатий
for key in p:
    press_count[key - 1] += 1  # Увеличиваем счетчик для соответствующей клавиши

# Проверка, какие клавиши сломались
result = []
for i in range(n):
    if press_count[i] > c[i]:
        result.append("YES")  # Клавиша сломалась
    else:
        result.append("NO")  # Клавиша работает
        
# Вывод результата
output = '\n'.join(result)
try:
    with open('output.txt', 'w') as f:
        f.write(output)
except:
    print(output)
```

```python
# Чтение входных данных
try:
    with open('input.txt', 'r') as f:
        text = f.read()
except FileNotFoundError:
    text = input()

text = text.split('\n')
c = [tuple(map(int,i.split())) for i in text[1:-1]]

blocks = dict()
for el in c:
    try:
        blocks[el[0]] = max(el[1],blocks[el[0]])
    except:
        blocks[el[0]] = el[1]
keys = (sorted(blocks.keys(),reverse=True))

output = str(sum([blocks[el] for el in keys]))+'\n'

try:
    with open('output.txt', 'w') as f:
        f.write(output)
except:
    print(output)
```

```python
# Чтение входных данных
try:
    with open('input.txt', 'r') as f:
        text = f.read()
except FileNotFoundError:
    text = input()

text = text.split('\n')
c = [tuple(i.split()) for i in text[:-1]]

people = dict()

for el in c:
    if (el[0] in people.keys()):
        if el[1] in people[el[0]].keys():
            people[el[0]][el[1]] += int(el[2])
        else:
            people[el[0]][el[1]] = int(el[2])
    else:
        people[el[0]] = {el[1]:int(el[2])}
        
result=[]
for name in sorted(people):
    result.append(f'{name}:')
    for el in sorted(people[name].keys()):
        result.append(f'{el} {people[name][el]}')

output = '\n'.join(result)
try:
    with open('output.txt', 'w') as f:
        f.write(output)
except:
    print(output)
```

```python
# Чтение входных данных
try:
    with open('input.txt', 'r') as f:
        text = f.read()
except FileNotFoundError:
    text = input()

text = text.split('\n')
c = [tuple(i.split()) for i in text]

people = dict()
result = []
for el in c:
    if not len(el):
        pass
    elif el[0] == 'DEPOSIT':
        if (el[1] in people.keys()):
            people[el[1]]+=int(el[2])
        else:
            people[el[1]]=int(el[2])
    
    elif el[0] == 'INCOME':
        for name in people:
            people[name] = people[name]*(100 + int(el[1]))//100 if people[name]>0 else people[name]
    
    elif el[0] == 'BALANCE':
        if (el[1] in people.keys()):
            result.append(str(people[el[1]]))
        else:
            result.append('ERROR')
            
    elif el[0] == 'TRANSFER':
        if (el[1] in people.keys()):
            people[el[1]]-=int(el[3])
        else:
            people[el[1]]= - int(el[3])
        if (el[2] in people.keys()):
            people[el[2]]+=int(el[3])
        else:
            people[el[2]]=int(el[3])
    
    elif el[0] == 'WITHDRAW':
        if (el[1] in people.keys()):
            people[el[1]]-=int(el[2])
        else:
            people[el[1]]= -int(el[2])

output = '\n'.join(result)

try:
    with open('output.txt', 'w') as f:
        f.write(output)
except:
    print(output)
```

```python
from collections import Counter

# Чтение входных данных
try:
    with open('input.txt', 'r') as f:
        text = f.read()
except FileNotFoundError:
    text = input()

lines = text.splitlines()
g, n = map(int, lines[0].split())
W = lines[1]
S = lines[2]

wdict = dict(Counter(W))
window = dict(Counter(S[:g]))
count = 1 if window == wdict else 0

# Скользящее окно: обновляем только изменившиеся символы
for i in range(g, n):
    left_char = S[i - g]  # символ, который выходит из окна
    right_char = S[i]     # символ, который входит в окно


    # Уменьшаем количество символа, выходящего из окна
    window[left_char] -= 1
    if window[left_char] == 0:
        del window[left_char]  # удаляем ключ, чтобы сравнение было быстрее

    try:
        # Увеличиваем количество символа, входящего в окно
        window[right_char] += 1
    except:
        window[right_char] = 1
    # Если текущее окно совпадает по частотам с wdict, увеличиваем счётчик
    if window == wdict:
        count += 1

output = str(count)
try:
    with open('output.txt', 'w') as f:
        f.write(output)
except Exception:
    print(output)
```

```python
# Чтение входных данных
try:
    with open('input.txt', 'r') as f:
        text = f.read()
except FileNotFoundError:
    text = input()

def i(txt):
    txt = txt.split('\n')
    N = int(txt[0])
    vocab = dict()
    for i in range(N):
        inp = txt[i+1]
        if inp.lower() in vocab.keys():
            vocab[inp.lower()].append(inp)
        else:
            vocab[inp.lower()] = [inp]

    errors = 0
    inpt = txt[N+1].split()

    if len(inpt):
        for word in inpt:

            if word == word.lower():
                errors += 1
            elif sum([int(str.isupper(i)) for i in word]) > 1:
                errors += 1
            elif word.lower() in vocab.keys():
                if not word in vocab[word.lower()]:
                    errors+=1

    return errors


output = str(i(text))
try:
    with open('output.txt', 'w') as f:
        f.write(output)
except Exception:
    print(output)
```

```python
import re

# Чтение входных данных
try:
        with open('input.txt', 'r') as f:
            data = f.read()
except FileNotFoundError:
        data = input()
    
lines = data.splitlines()

# Первая строка содержит: число ключевых слов n, настройки чувствительности к регистру (C) и возможности начинать с цифры (D)
parts = lines[0].split()
n = int(parts[0])
case_sensitive = (parts[1] == "yes")
digit_start_allowed = (parts[2] == "yes")
    
# Считываем ключевые слова
keywords = set()
for i in range(1, 1 + n):
        kw = lines[i]
        if not case_sensitive:
            kw = kw.lower()
        keywords.add(kw)
    
# Остальные строки – текст программы
program_text = "\n".join(lines[1+n:])
    
# Используем регулярное выражение для выделения всех последовательностей, состоящих из букв, цифр и подчеркиваний
tokens = re.findall(r"[A-Za-z0-9_]+", program_text)
    
# Подсчёт встречаемости идентификаторов и запоминание порядка первого появления
counts = {}           # token -> количество вхождений
first_occurrence = {} # token -> порядковый номер первого появления
order = 0
    
for token in tokens:
        # Приводим токен к нужному регистру, если язык не чувствителен к регистру
        proc_token = token if case_sensitive else token.lower()
        
        # Если токен совпадает с ключевым словом, пропускаем его
        if proc_token in keywords:
            continue
        
        # Токен должен содержать хотя бы один символ, не являющийся цифрой
        if proc_token.isdigit():
            continue
        
        # Если идентификаторы не могут начинаться с цифры, отбрасываем такие токены
        if not digit_start_allowed and proc_token[0].isdigit():
            continue
        
        # Это допустимый идентификатор – увеличиваем счетчик
        if proc_token not in counts:
            counts[proc_token] = 0
            first_occurrence[proc_token] = order
            order += 1
        counts[proc_token] += 1

# Ищем идентификатор с максимальным числом вхождений; при равенстве выбираем тот, что встретился раньше
best_identifier = None
best_count = -1
best_order = float('inf')
for ident, cnt in counts.items():
        occ = first_occurrence[ident]
        if cnt > best_count or (cnt == best_count and occ < best_order):
            best_identifier = ident
            best_count = cnt
            best_order = occ

# Вывод результата
output = best_identifier if best_identifier is not None else ""
try:
        with open('output.txt', 'w') as f:
            f.write(output)
except Exception:
        print(output)
```
