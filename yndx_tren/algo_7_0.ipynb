# B

Массив `dp[i]` хранит минимальное время для обслуживания первых `i` человек. Для каждого `i`  рассматриваются три варианта группировки (1, 2 или 3 человека) и выбирается оптимальный

```python
try:
    with open('input.txt', 'r') as f:
        text = f.read().split('\n')
except FileNotFoundError:
    text = input().split('\n')
    
N = int(text[0])
arrays = list(map(lambda x : list(map(lambda x: int(x) ,x.split())),text[1:-1]))

A = [arrays[i][0] for i in range(N)]
B = [arrays[i][1] for i in range(N)]
C = [arrays[i][2] for i in range(N)]

# Инициализация динамического программирования
dp = [float('inf')] * (N + 3)
dp[0] = 0
for i in range(1, N + 1):
    # Вариант: текущий человек покупает 1 билет
    dp[i] = min(dp[i], dp[i-1] + A[i-1])
    # Вариант: предыдущий и текущий покупают 2 билета (если i >= 2)
    if i >= 2:
        dp[i] = min(dp[i], dp[i-2] + B[i-2])
    # Вариант: три человека покупают 3 билета (если i >= 3)
    if i >= 3:
        dp[i] = min(dp[i], dp[i-3] + C[i-3])

output = str(dp[N])

try:
    with open('output.txt', 'w') as f:
        f.write(output)
except:
    print(output)
```

# C

Для решения задачи требуется соединить гвоздики ниточками так, чтобы каждый гвоздик был привязан к хотя бы одной ниточке, а суммарная длина всех ниточек была минимальной. Оптимальный подход заключается в использовании динамического программирования, где мы рассматриваем разбиение гвоздиков на группы по два или три подряд для минимизации суммарной длины.

- `dp[i]` — минимальная суммарная длина ниточек для первых `i` гвоздиков.
- `dp[0] = 0`: Нулевой гвоздиков не требуют ниточек.
- `dp[1] = бесконечность`: Один гвоздик невозможно покрыть (нужна хотя бы пара), поэтому это значение недостижимо .

- **Для каждого `i` (от 2 до `n`):**
  - **Вариант 1:** Соединить `(i-1)`-й и `i`-й гвоздики (группа из 2). Длина ниточки — разница их координат. К сумме добавляется `dp[i-2]` (минимальная длина для предыдущих `i-2` гвоздиков).
  - **Вариант 2:** Соединить `(i-2)`-й, `(i-1)`-й и `i`-й гвоздики (группа из 3). Длина ниточки — разница между крайними гвоздиками в группе. К сумме добавляется `dp[i-3]` (минимальная длина для предыдущих `i-3` гвоздиков) .

Значение `dp[n]` содержит минимальную суммарную длину ниточек для всех `n` гвоздиков, учитывая все возможные комбинации групп по 2 или 3 подряд .

```python
try:
    with open('input.txt', 'r') as f:
        text = f.read().split('\n')
except FileNotFoundError:
    text = input().split('\n')
    
n = int(text[0])

x = sorted(list(map(int,text[1].split())))
print(x)
dp = [0] * (n + 1)
dp[0] = 0
if n >= 1:
    dp[1] = float('inf')

for i in range(2, n + 1):
    option1 = dp[i-2] + (x[i-1] - x[i-2])
    option2 = float('inf')
    if i >= 3:
        option2 = dp[i-3] + (x[i-1] - x[i-3])
    dp[i] = min(option1, option2)
    
    print(f'\ncurr pin = {x[i-1]}')
    print('options:')
    print(f'{dp[i-2]} + ({x[i-1]} - {x[i-2]}) = {option1}')
    if i >= 3:
        print(f'{dp[i-3]} + ({x[i-1]} - {x[i-3]}) = {option2}')
    
    

output = str(dp[n])

print(f'\n\nOutput = {output}')
try:
    with open('output.txt', 'w') as f:
        f.write(output)
except:
    print(output)
```

## OK решение

```python
try:
    with open('input.txt', 'r') as f:
        text = f.read().split('\n')
except FileNotFoundError:
    text = input().split('\n')

n = int(text[0])
prices = list(map(int, text[1:-1]))

# dp[i][c] = (min_cost, final_coupons, decision)
# decision: 0 - оплата наличными, 1 - использование купона.
# Для восстановления решения храним выбор для состояния (i, c)
# Так как максимальное число купонов, которое может накопиться, не больше n,
# создадим таблицу размером (n+1) x (n+1)
dp = [[(10**9, -10**9, -1) for _ in range(n+2)] for _ in range(n+1)]

# База: в день n (после всех дней) стоимость = 0, купонов = c (накапливаются)
for c in range(n+2):
    dp[n][c] = (0, c, -1)

# Обратный перебор по дням
for i in range(n-1, -1, -1):
    for c in range(n+1):
        # Опция 1: использование купона (если доступно)
        best = (10**9, -10**9, -1)
        if c > 0:
            cost_coupon, final_coupons_coupon, _ = dp[i+1][c-1]
            best = (cost_coupon, final_coupons_coupon, 1)  # 1 означает "использовать купон"
        
        # Опция 2: оплата наличными
        p = prices[i]
        new_c = c + (1 if p > 100 else 0)
        cost_pay, final_coupons_pay, _ = dp[i+1][new_c]
        cost_pay += p
        
        # Выбор: минимизируем стоимость, а при равенстве максимизируем итоговое число купонов
        option_pay = (cost_pay, final_coupons_pay, 0)  # 0 означает "оплатить"
        
        # Сравниваем best и option_pay
        if option_pay[0] < best[0] or (option_pay[0] == best[0] and option_pay[1] > best[1]):
            best = option_pay
        dp[i][c] = best

# Восстановление решения
used_days = []
coupons = 0
i = 0
while i < n:
    decision = dp[i][coupons][2]
    if decision == 1:
        # Использование купона
        used_days.append(i+1)  # дни нумеруются с 1
        coupons -= 1
    else:
        # Оплата наличными
        p = prices[i]
        if p > 100:
            coupons += 1
    i += 1

total_cost, final_coupons, _ = dp[0][0]
used_count = len(used_days)

out = []
out.append(str(total_cost))
out.append(f"{final_coupons} {used_count}")
for day in used_days:
    out.append(str(day))
    
output = "\n".join(out)

try:
    with open('output.txt', 'w') as f:
        f.write(output)
except:
    print('\n'.join(output))
```

```python
class Heap:
    def __init__(self):
        self.heap = []  # список для хранения элементов кучи

    def Insert(self, k):
        """Добавляет число k в кучу."""
        self.heap.append(k)
        self._sift_up(len(self.heap) - 1)

    def Extract(self):
        """Извлекает и возвращает наибольшее число из кучи."""
        if not self.heap:
            raise Exception("Heap is empty")
        max_value = self.heap[0]
        # Заменяем корень последним элементом
        last = self.heap.pop()
        if self.heap:
            self.heap[0] = last
            self._sift_down(0)
        return max_value

    def _sift_up(self, index):
        """Поднимает элемент на позиции index вверх до восстановления свойства кучи."""
        while index > 0:
            parent = (index - 1) // 2
            if self.heap[index] > self.heap[parent]:
                self.heap[index], self.heap[parent] = self.heap[parent], self.heap[index]
                index = parent
            else:
                break

    def _sift_down(self, index):
        """Опускает элемент с позиции index вниз до восстановления свойства кучи."""
        size = len(self.heap)
        while True:
            left = 2 * index + 1
            right = 2 * index + 2
            largest = index

            if left < size and self.heap[left] > self.heap[largest]:
                largest = left
            if right < size and self.heap[right] > self.heap[largest]:
                largest = right

            if largest != index:
                self.heap[index], self.heap[largest] = self.heap[largest], self.heap[index]
                index = largest
            else:
                break

try:
    with open('input.txt', 'r') as f:
        text = f.read().split('\n')
except FileNotFoundError:
    text = input().split('\n')

n = int(text[0])
comands = list(map(lambda x : list(map(lambda x: int(x) ,x.split())),text[1:-1]))


heap = Heap()
out = []
for i in comands:
    if i[0] == 0:
        heap.Insert([i[1]])
    else:
        out.append(str(heap.Extract()[0]))
        
    
output = "\n".join(out)

try:
    with open('output.txt', 'w') as f:
        f.write(output)
except:
    print('\n'.join(output))
```

```python
try:
    with open('input.txt', 'r') as f:
        text = f.read().split('\n')
except FileNotFoundError:
    text = input().split('\n')

N = int(text[0])
lines = list(map(lambda x: x.split(), text[1:-1]))

parents = {}
for i in range(N - 1):
    child = lines[i][0]
    parent = lines[i][1]
    parents[child] = parent

# Найти родоначальника
all_children = set(parents.keys())
all_parents = set(parents.values())
root = (all_parents - all_children).pop()

heights = {root: 0}

# Обрабатываем всех
all_names = all_children.union(all_parents)

for name in all_names:
    if name not in heights:
        path = []
        current = name
        while current not in heights:
            path.append(current)
            current = parents[current]
        h = heights[current]
        while path:
            node = path.pop()
            h += 1
            heights[node] = h

out=[]
# Сортируем и выводим
sorted_names = sorted(heights.keys())
for name in sorted_names:
    out.append(f"{name} {heights[name]}")
    
output = "\n".join(out)

try:
    with open('output.txt', 'w') as f:
        f.write(output)
except:
    print('\n'.join(output))
```

```python
# Бинарное дерево
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def add(self, data):
        if not (self.search(data)):
            new_node = Node(data)
            if self.root is None:
                self.root = new_node
            else:
                current = self.root
                while True:
                    if data < current.data:
                        if current.left is None:
                            current.left = new_node
                            break
                        else:
                            current = current.left
                    else:
                        if current.right is None:
                            current.right = new_node
                            break
                        else:
                            current = current.right
            return True
        else:
            return False

    def search(self, data):
        current = self.root
        while current is not None:
            if data == current.data:
                return True
            elif data < current.data:
                current = current.left
            else:
                current = current.right
        return False

    

    def __str__(self):
        lines = []
        self._in_order(self.root, 0, lines)
        
        return lines #'\n'.join(lines)
    
    def _in_order(self, node, depth, lines):
        if node is None:
            return
        # Сначала обрабатываем левое поддерево
        self._in_order(node.left, depth + 1, lines)
        # Добавляем текущий узел с нужным количеством точек
        lines.append('.' * depth + str(node.data))
        # Затем обрабатываем правое поддерево
        self._in_order(node.right, depth + 1, lines)
    


try:
    with open('input.txt', 'r') as f:
        text = f.read().split('\n')
except FileNotFoundError:
    text = input().split('\n')

tree = BinaryTree()
comands = list(map(lambda x: x.split(), text[:-1]))
out=[]
for com in comands:
    if com[0] == 'ADD':
        if tree.add(int(com[1])):
            out.append('DONE')
        else:
            out.append('ALREADY')
    elif com[0] == 'SEARCH':
        if tree.search(int(com[1])):
            out.append('YES')
        else:
            out.append('NO')
    elif com[0] == 'PRINTTREE':
        lines = []
        tree._in_order(tree.root, 0, lines)
        out.extend(lines)
    
output = "\n".join(out)

try:
    with open('output.txt', 'w') as f:
        f.write(output)
except:
    print('\n'.join(output))
```

```python
from collections import deque

try:
    with open('input.txt', 'r') as f:
        text = f.read().split('\n')
except FileNotFoundError:
    text = input().split('\n')

def depth(conns,val):
    s = 1
    try:
        for i in conns[val]:
            s+= depth(conns,i)
    except:
        s = 1
    return s
    

n = int(text[0])
edges = list(map(lambda x: list(map(int,x.split())), text[1:-1]))



# Строим список смежности
adj = [[] for _ in range(n + 1)]
for u, v in edges:
    adj[u].append(v)
    adj[v].append(u)


# Определяем детей для каждой вершины, начиная с корня (1)
parent = [0] * (n + 1)
children = [[] for _ in range(n + 1)]
q = deque([1])
parent[1] = -1  # Корень не имеет родителя

while q:
    u = q.popleft()
    for v in adj[u]:
        if parent[v] == 0:
            parent[v] = u
            children[u].append(v)
            q.append(v)
            
# Вычисляем размеры поддеревьев через пост-порядковый обход
size = [1] * (n + 1)
stack = [(1, False)]
post_order = []

while stack:
    node, visited = stack.pop()
    if visited:
        post_order.append(node)
        continue
    stack.append((node, True))
    # Добавляем детей в обратном порядке для правильной обработки
    for child in reversed(children[node]):
        stack.append((child, False))

# Обновляем размеры поддеревьев
for u in post_order:
    for child in children[u]:
        size[u] += size[child]

# Формируем результат
output = ' '.join(map(str, size[1:n+1]))

try:
    with open('output.txt', 'w') as f:
        f.write(output)
except:
    print('\n'.join(output))
```
